```python
import requests
import json
import os


# 使用方法
url = 'https://www.zoomeye.org/api/search?q=title%3A%22%E5%B0%8F%E9%9B%85%E7%9A%84%E5%88%86%E7%B1%BB+Alist%22+%2Bcountry%3ACN+%2Bsubdivisions%3A%E5%B9%BF%E4%B8%9C+%2Bapp%3A%22nginx%22&page=1&t=v4%2Bv6%2Bweb%2Bhost'

# 文件路径
file_path = 'ip.json'


def get_ips_from_zoomeye(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36',
        'Accept': 'application/json, text/plain, */*'
    }

    response = requests.get(url, headers=headers)

    # 检查请求是否成功
    if response.status_code == 200:
        try:
            # 解析JSON数据
            data = response.json()
            # 提取"ips"属性
            ips = data.get('ips', [])
            return ips
        except json.JSONDecodeError:
            print("响应内容不是有效的JSON格式")
            return []
    else:
        print(f"请求失败，状态码: {response.status_code}")
        return []

def test_api(ip_address):
    url = f"http://{ip_address}:5678/api/fs/get"
    headers = {
        "accept": "application/json, text/plain, */*",
        "accept-language": "zh-CN,zh;q=0.9,en;q=0.8,en-GB;q=0.7,en-US;q=0.6",
        "authorization": "",
        "content-type": "application/json;charset=UTF-8"
    }
    body = {
        "path": "/动漫/儿童/BBC 字母积木 1-4季/积木英语Alphablocks第一季【598.78MB】/Alphablocks_-_1._Alphablocks.avi",
        "password": ""
    }

    try:
        # 发送HTTP POST请求
        response = requests.post(url, headers=headers, json=body, allow_redirects=True)

        # 检查请求是否成功
        if response.status_code == 200:
            try:
                # 尝试解析JSON数据
                data = response.json()
                # 检查响应是否包含code字段，并且code的值是否为401
                if 'code' in data and data['code'] == 401:
                    print("Guest user is disabled, login please")
                    return False
                else:
                    #print(json.dumps(data, indent=4, ensure_ascii=False))
                    return True
            except json.JSONDecodeError:
                print("响应内容不是有效的JSON格式")
                return False
        else:
            print(f"请求失败，状态码: {response.status_code}")
            return False
    except requests.RequestException as e:
        print(f"请求异常: {e}")
        return False
    
    
def check_and_clean_ip_list(ip_list):
    # 创建一个新的列表来存储通过测试的IP地址
    valid_ips = []

    # 遍历原始IP列表
    for ip in ip_list:
        # 调用 test_api 函数来检查IP是否通过测试
        if test_api(ip):
            # 如果通过测试，则添加到新的列表中
            valid_ips.append(ip)
        else:
            # 如果未通过测试，则打印一条消息并继续下一个IP
            print(f"IP {ip} failed the test.")

    # 返回通过测试的IP列表
    return valid_ips
    
    
def read_ips_from_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
            return data.get('ips', [])
    except FileNotFoundError:
        return []
    except json.JSONDecodeError:
        print("JSON文件格式错误")
        return []

    
def write_ips_to_file(file_path, ip_list):
    with open(file_path, 'w', encoding='utf-8') as file:
        json.dump({'ips': ip_list}, file, indent=4, ensure_ascii=False) 

        
ip_list_json=read_ips_from_file(file_path)

        
ip_list_zoomeye = get_ips_from_zoomeye(url)
print("获得新ip")
print(ip_list_zoomeye)
merged_ips = list(set(ip_list_json + ip_list_zoomeye))
cleaned_ip_list = check_and_clean_ip_list(merged_ips)


print(cleaned_ip_list)
```
